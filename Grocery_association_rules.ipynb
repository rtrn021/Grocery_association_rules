```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df=pd.read_csv("data\GroceryStoreDataSet.csv",header=None)
```

```python
df
```

```python
items=[]
for i in df.values:
    items.extend( i[0].split(",")  )
items=list(set(items))
items
```

```python
df1=pd.DataFrame(data=0,columns=items,index=range(len(df)))
df1
```

```python
for i in df1.columns:
    df1[i] = df[0].str.contains(i)
#     df1[i]=a[0].apply(lambda x: int(i in x)) >> to change to numeric
df1

# Another way To Create with 1 and 0
# row=0
# for i in df.values:
#     for j in df1.columns:
#         if j in i[0].split(","):
#             df1[j][row]=1
#     row+=1
```

```python
freq=apriori(df1, min_support = 0.2, use_colnames = True)
```

```python
freq
```

```python
association_rules(freq, metric = "lift", min_threshold = 1).sort_values(by=['antecedent support','consequent support'],ascending=False).reset_index(drop=True)
```
